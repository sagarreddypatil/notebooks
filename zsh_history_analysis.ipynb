```python
import os
history = open(os.path.expanduser("~/.zsh_history"), "r", encoding='latin-1').read()

history = ("\n" + history).split("\n: ")[1:]
```

```python
import matplotlib.dates as mdates
from datetime import datetime
from time import time

def gettime(line):
    return int(line.split(":")[0])

now = int(time())

times = map(gettime, history)
times = filter(lambda x: x <= now, times)
times = map(lambda x: datetime.fromtimestamp(x), times)
times = list(times)
days = [x.date() for x in times]
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, ax = plt.subplots()
ax.hist(days, bins=100, color='b')
# ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate()
plt.show()
```
